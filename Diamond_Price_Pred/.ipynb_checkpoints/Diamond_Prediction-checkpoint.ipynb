*Importing Libraries*

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
from out_r import replace_outliers
```

*Loading Daimond Dataset*

```python
df = pd.read_csv('diamonds.csv')
df.drop('Unnamed: 0',axis=1,inplace=True)
df.head()
```

```python
df.describe().T
```

```python
df_ob=df.select_dtypes(include = ['object'])
cat_col = (df_ob.nunique()).index.tolist()
df.drop(cat_col,axis=1).skew()
```

# EDA

```python
sns.boxplot(df.drop('price',axis=1))
```

### Removing Outliers 

```python
df_con_col = df.drop(columns=cat_col,axis=1)
df_cat_col = df.drop(columns=df_con_col.columns.to_list(),axis=1)
df = df_cat_col.join([replace_outliers(df_con_col)])
plt.figure(figsize=(20,10))
sns.boxplot(df)
```

# Preprocessing 

*Hot Encoding of Categorical Columns*

```python
encoded_df = pd.get_dummies(df, columns=cat_col)
pd.set_option('display.max_columns', 50)
encoded_df.head()
```

# Model Training 

```python
from sklearn.model_selection import train_test_split , KFold, cross_val_score
y = encoded_df['price']
X = encoded_df.drop('price',axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
Lr = LinearRegression()
Lr.fit(X_train,y_train)
```

```python
cv = KFold(n_splits=20, random_state=42, shuffle=True)

# Use k-fold CV to evaluate model (mean absolute error)
scores = cross_val_score(Lr, X, y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
mean_absolute_error = np.mean(np.abs(scores))
print(f"Mean Absolute Error: {mean_absolute_error:.4f}")
```

```python
predictions = np.abs(Lr.predict(X_test) )
```

# Evaluating Predictions 

```python
plt.scatter(y_test,predictions,edgecolors='white')
plt.xlabel('Actual values')
plt.axvline(c='grey', lw=1)
plt.axhline(c='grey', lw=1)
plt.ylabel('Predictions')
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
